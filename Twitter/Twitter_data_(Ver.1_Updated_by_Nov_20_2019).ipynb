## Access Twitter API

```python
with open("twitter_credentials.json", "r") as file:
    creds = json.load(file)
```

```python
auth = tw.OAuthHandler(creds['CONSUMER_KEY'], creds['CONSUMER_SECRET'])
auth.set_access_token(creds['ACCESS_TOKEN'], creds['ACCESS_SECRET'])
api = tw.API(auth, wait_on_rate_limit=True)
```

```python
terms_df = pd.read_csv(r"D:\Yeshiva_Univ\Courses\Fall_2019\Informational_Archtecture\Final_Project\terms_hatebase.csv")

terms_df.head()
```

```python
#Pullout only the hate term variables
hateterms = terms_df['term']

hateterms
```

```python
#Add hash tags to the infront of the words. 
hateterms =  '#' + hateterms.astype(str)

hateterms.head()
```

```python
# Change the datatype of hateterms from a series to a list
hateterms_1 = hateterms.values.tolist()

hateterms_1
```

```python
# Use For loop function  
search_words = hateterms
date_since = '2019-01-01'

list = [] #
for i in hateterms_1:
    tweets = tw.Cursor(api.search,
                       q=i,
                       lang="en",
                       since=date_since).items(200)
    list.extend([tweet.text for tweet in tweets])
```
